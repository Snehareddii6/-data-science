```python
#difference
a = {1,4,5,7,9}
b = {2,4,5,7,11}
print(a.difference(b))
#print(b.difference(a))
```

```python
#intersection
x={1,2,3,4,5,6}
y={8,7,6,5,4,3}
intersection_set = x.intersection(y)
print(intersection_set)
```

```python
#symmetric_difference
x={1,2,3,4,5}
y={3,4,5,6,7}
result= x.symmetric_difference(y)
print(result)
```

```python
#union
X={1,2,3,5}
Y={1,3,5,6}
print('X U Y=',X.union(Y))
```

```python
#issuperset
X = {1,2,3,4,5,6}
Y = {1,2,3,4}
Z = {1,2,3,4}
print(X.issuperset(Y))
print(Z.issuperset(X))
```

```python
#Iissubset
x={1,2,3}
y={1,2,3,4,5}
print(x.issubset(y))
```

```python
#isdisjoint
X={4,5,6,7}
Y={8,9,10}
print(X.isdisjoint(Y))
```
